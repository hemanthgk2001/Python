```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('medicines.csv')
df1 = pd.DataFrame(df)
```

```python
# displaying first five elements
print(df1.head())
```

```python
# displaying last five elements
print(df1.tail())
```

```python
# displays the information about the dataset
print(df1.info())
```

```python
# displays the description of the dataset
print(df1.describe())
```

```python
# displays the shape of the dataframe
print(df1.shape)
```

```python
# checking for the sum of null values
df1.isnull().sum()
```

```python
# checking for the sum of not null values
df1.notnull().sum()
```

```python
# to sort the values in decending order
m = df1.sort_values('price(₹)',ascending=False)
print(m.head(6))
```

```python
# to drop one column
x = df1.drop(['pack_size_label'],axis=1).head(5)
print(x)
```

```python
# to display the columns in the dataset
disp = df1.columns
print(disp)
```

```python
# to find the length of overall data
l = len(df1.index.value_counts())
print(l)
```

```python
# displaying the filtering function ,extracting from last 20 to 10 rows and from 1 to 7 columns in reverse order
disp1 = df1.iloc[-20:-10,7:1:-1]
# plotting the sns relational plot
p = sns.relplot(x='price(₹)',y='pack_size_label',data=df1.iloc[0:4])
q = sns.relplot(x='price(₹)',y='pack_size_label',data=df1.iloc[0:4],kind='line',aspect=2)
print(p)
print(q)
```

```python
# plotting the Manga and Sales by using histograph
plot1 = plt.figure(figsize=(10,8))
plot2 = plt.hist(df1['price(₹)'].head(5))
plt.show()
```

```python
# bar graph
plt.figure(figsize=(12,8))
plt.bar(df1['name'].head(5),df1['price(₹)'].head(5))# use bar plot for categorical vs numerical data
plt.xlabel('name')
plt.ylabel('price(₹)')
plt.title("TOP 5 MEDICINE AND ITS SALES")
plt.show()
```

```python
# linesplot
plt.figure(figsize=(10,8))
sns.lineplot(x='name', y='type' ,data = df1.head(5))# remove hue and style
plt.xlabel('name')
plt.ylabel('type')
plt.title("TOP 5 MEDICINE AND ITS SALES")
plt.show()
```
